# **Modelo Ataque do Coração**

```python
import pandas as pd
import numpy as np
```

```python
base_cor = pd.read_csv("heart_attack_prediction_dataset.csv")
```

```python
base_cor.columns
```

```python
base_cor
```

```python
base_cor.isna().sum()
```

```python
base_cor.info()
```

```python
categoricas = ["Sex","Diet","Blood Pressure","Country","Continent","Hemisphere"]#BMI pode ser importante
```

```python
base_cor.drop(columns="Patient ID", inplace=True)
```

```python
#base_cor_dumm = pd.get_dummies(base_cor, columns=categoricas, drop_first=True)
```

```python
base_cor_dumm = base_cor.drop(columns=categoricas)
```

```python
from sklearn.preprocessing import MinMaxScaler
norm = MinMaxScaler()
base_cor_dumm_norm = pd.DataFrame(norm.fit_transform(base_cor_dumm), columns=base_cor_dumm.columns)
```

```python
base_cor_dumm_norm = base_cor_dumm
```

# **Divisão Treino/Teste, Alg. M.L., Validação, Normalização**

```python
from sklearn.model_selection import train_test_split

y = base_cor_dumm_norm["Heart Attack Risk"]
X = base_cor_dumm_norm.drop(columns="Heart Attack Risk")
#y = base_cor_dumm["Heart Attack Risk"]
#X = base_cor_dumm.drop(columns="Heart Attack Risk")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#from sklearn.naive_bayes import GaussianNB
#from sklearn.tree import DecisionTreeClassifier
```

```python
modelo = KNeighborsClassifier(n_neighbors=3)
#modelo = GaussianNB()
#modelo = DecisionTreeClassifier()
```

```python
modelo.fit(X_train,y_train)
```

```python
modelo.score(X_train,y_train)
```

```python
pred = modelo.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
resp = pd.DataFrame(y_test)
resp["pred"] = pred
resp
```
